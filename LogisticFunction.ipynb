```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the logistic function
k = .7  # Steepness of the curve

# Generate x values
x = np.linspace(0, 20, 400)

# Calculate the corresponding y values using the modified logistic function
y = (1 / (1 + np.exp(k * (x - np.max(x)/2))))

# Plot the logistic function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Logistic Function', color='b')
plt.title('Logistic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```
